```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = cv2.imread('images/Sunflower.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(image_copy)
```

```python
#convert to gray scale
gray_img = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(gray_img, cmap='gray')
```

```python
#canny edge detector
#lower and upper threshold for hysteresis
lower = 180
upper = 240    #ratio 1:2

#Canny produces binary image with thin lines
edge_img = cv2.Canny(gray_img,lower,upper)

plt.figure(figsize=(20,10))
plt.imshow(edge_img, cmap='gray')
```

```python
#Canny using wide and tight threshold
wide_img = cv2.Canny(gray_img,30,100)      #lower threshold
tight_img = cv2.Canny(gray_img,180,240)    #narrow range for exceptable edges

f,(plt1,plt2) = plt.subplots(1,2,figsize=(20,10))

plt1.set_title('wide img')
plt1.imshow(wide_img, cmap='gray')

plt2.set_title('tight img')
plt2.imshow(tight_img, cmap='gray')
```

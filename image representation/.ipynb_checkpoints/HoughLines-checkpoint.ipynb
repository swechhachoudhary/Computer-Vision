```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = cv2.imread('images/hand-holding-iphone.jpg')

img = np.copy(image)

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img)
```

```python
#convert to gray scale image and apply canny edge detection
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

#thresholds
low = 50
upper = 100
edge_img = cv2.Canny(gray,low,upper)

plt.figure(figsize=(10,10))
plt.imshow(edge_img, cmap='gray')
```

```python
#find lines using Hough Transform
rho = 1   # 1 pixel
theta = np.pi/180   # 1 degree
threshold = 60     #minimum no of hough space intersection to define a line
min_line_leng = 70
max_line_gap = 5

lines = cv2.HoughLinesP(edge_img,rho,theta,threshold,np.array([]),min_line_leng,max_line_gap)
#each line is an array of 4 points

line_img = np.copy(img)

for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_img, (x1,y1), (x2,y2), (255,0,0), 2)

plt.figure(figsize=(10,10))
plt.imshow(line_img)  #phone screen if detected
```

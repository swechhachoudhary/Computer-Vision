```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
image = cv2.imread('images/pizza2.jpg')
print('This image is', type(image), 
      'with dimensions ', image.shape)
```

```python
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

 Color Threshold

```python
lower_blue = np.array([10,10,200])
upper_blue = np.array([200,200,255])

#create image mask
mask = cv2.inRange(image_copy,lower_blue,upper_blue)

plt.imshow(mask, cmap= 'gray')
```

```python
#mask the image to let the pizza show through
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0,0,0]
plt.imshow(masked_image)
```

```python
background_img = cv2.imread('images/space.jpg')
image_copy = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)
crop_back_img = background_img[0:335,0:596]
crop_back_img[mask == 0] = [0,0,0]
plt.imshow(crop_back_img)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
image = cv2.imread('images/balloon.jpg')
print('This image is', type(image), 
      'with dimensions ', image.shape)
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
r = image_copy[:,:,0]
g = image_copy[:,:,1]
b = image_copy[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r,cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')
```

```python
hsv_img = cv2.cvtColor(image_copy,cv2.COLOR_RGB2HSV)

h = hsv_img[:,:,0]
s = hsv_img[:,:,1]
v = hsv_img[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Value')
ax3.imshow(v,cmap='gray')
```

```python
lower_pink = np.array([180,0,100])
upper_pink = np.array([255,255,230])

lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255]) #max hue value is 180

#create image mask
RGB_mask = cv2.inRange(image_copy,lower_pink,upper_pink)

plt.imshow(RGB_mask, cmap= 'gray')
```

```python
masked_image = np.copy(image_copy)
masked_image[RGB_mask == 0] = [0,0,0]
plt.imshow(masked_image)
plt.title('RGB selection image')
```

```python
#create image mask for hsv
HSV_mask = cv2.inRange(hsv_img,lower_hue,upper_hue)

plt.imshow(HSV_mask, cmap= 'gray')
```

```python
#HSV masked image
HSVmasked_image = np.copy(image_copy)
HSVmasked_image[HSV_mask == 0] = [0,0,0]
plt.imshow(HSVmasked_image)
plt.title('HSV selection image')
```
